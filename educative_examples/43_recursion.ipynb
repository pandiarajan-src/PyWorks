```python
# Product of Two Positive Integers using recursive not using "*"
def recursive_multiply(x, y):
    # This helps to reduce the number of recursion
    if x < y:
        return recursive_multiply(y, x)
    if y <= 0:
        return 0
    return x + recursive_multiply(x, y-1)

# Fibonacci
def fibonacci(n):
    if n==1:
        return 0
    elif n==2:
        return 1
    else:
        return fibonacci(n-1)+fibonacci(n-2) 

# Find Uppercase Letter in String
def find_upper_case_letter_iterative(data):
    for letter in data:
        if letter.isupper():
            return letter
    return "No upper case letter found"


def find_uper_case_letter_recursive(data, id=0):
    if id > len(data)-1:
        return "No upper case letter found"
    if data[id].isupper():
        return data[id]
    id += 1
    return find_uper_case_letter_recursive(data, id)

# Find length of a string
def find_string_length_iterative(data):
    length = 0
    for letter in data:
        length += 1
    return length

def find_string_length_recursive(data):
    if data == '':
        return 0
    return 1 + find_string_length_recursive(data[1:])

# Find constants in strings
def count_constants_in_string(data):
    constants = "aeiou"
    count = 0 
    for letter in data:
        if letter.lower() not in constants and letter.isalpha():
            count += 1
    return count

def count_constants_in_string_recursive(data):
    constants = "aeiou"
    if data == '':
        return 0
    if data[0].lower() not in constants and data[0].isalpha():
        return 1 + count_constants_in_string_recursive(data[1:])
    else:
        return count_constants_in_string_recursive(data[1:])

str1 = "lucidProgramming"
str2 = "LucidProgramming"
str3 = "lucidprogramming"

print(f"Upper case in {str1} is : {find_upper_case_letter_iterative(str1)}")
print(f"Upper case in {str1} is : {find_uper_case_letter_recursive(str1)}")

print(f"Upper case in {str2} is : {find_upper_case_letter_iterative(str2)}")
print(f"Upper case in {str2} is : {find_uper_case_letter_recursive(str2)}")

print(f"Upper case in {str3} is : {find_upper_case_letter_iterative(str3)}")
print(f"Upper case in {str3} is : {find_uper_case_letter_recursive(str3)}")

print(f"Lenght of '{str1}' is : {find_string_length_iterative(str1)}")
print(f"Lenght of '{str1}' is : {find_string_length_recursive(str1)}")

print(f"Number of constonants in {str1} is : {count_constants_in_string(str1)}")
print(f"Number of constonants in {'abc de'} is : {count_constants_in_string('abc de')}")
print(f"Number of constonants in {str1} is : {count_constants_in_string_recursive(str1)}")
print(f"Number of constonants in {'abc de'} is : {count_constants_in_string_recursive('abc de')}")

print(fibonacci(7)) 

print(recursive_multiply(5, 3))
print(recursive_multiply(500, 2000))
```

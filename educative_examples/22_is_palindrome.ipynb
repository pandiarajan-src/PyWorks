# IS Palindrome

## What is Palindrome?
A palindrome is a word or a sequence that is the same from the front as from the back. 
For instance, racecar and radar are palindromes. 
Examples of non-palindromes are ABC, hello, and test.

### Metods to explore
1. Using String comparision (more pythonic way)
2. Using Stack method
3. Using two pointers comparision

```python
# IS Palindrome???

from singly_linked_list import LinkedList
from stack import Stack

def prepare_link_list(data):
    llst = LinkedList()
    data = list(data)
    for d in data:
        llst.append(d)
    return llst

def is_palindrome_by_string_logic(llst):
    curr_node = llst.head
    str_data = ""
    while curr_node:
        str_data += curr_node.data
        curr_node = curr_node.next
    return str_data == str_data[::-1]

def is_palindrome_by_stack_logic(llst):
    curr_node = llst.head
    stack_obj = Stack()
    while curr_node:
        stack_obj.push(curr_node.data)
        curr_node = curr_node.next

    curr_node = llst.head
    while curr_node:
        if curr_node.data != stack_obj.pop():
            return False
        curr_node = curr_node.next
    return True

def is_palindrom_by_2pointer_logic(llst):
    curr_node = llst.head
    if curr_node:
        rev = []
        while curr_node:
            rev.append(curr_node)
            curr_node = curr_node.next
        
        length = len(rev)
        count = 0
        curr_node = llst.head
        while count < (length//2 + 1):
            if curr_node.data != rev[-1].data:
                return False
            count += 1
        return True
    else:
        return True


def test_palindrome(input):
    print(f"Given input : {input}")
    print("Input as Singly linked list\n---")
    lst = prepare_link_list(input)
    lst.print_list()
    print("---")
    print(f"Is '{input}' Palindrome??? By String Method : {is_palindrome_by_string_logic(lst)}")
    print(f"Is '{input}' Palindrome??? By Stack Method : {is_palindrome_by_stack_logic(lst)}") 
    print(f"Is '{input}' Palindrome??? By 2 pointers in linked list Method : {is_palindrom_by_2pointer_logic(lst)}") 

    print("****************************")



test_palindrome("radar")
test_palindrome("racecar")
test_palindrome("hello")
test_palindrome("")
```

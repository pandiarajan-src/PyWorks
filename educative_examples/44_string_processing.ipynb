```python
# Look and Say sequence
# Generate the next number based on previous number

# For example, 1 is read off as one 1 which implies that the count of 1 is one. 
# As 1 is read off as “one 1”, the next sequence will be written as 11 where 1 replaces one. 
# Now 11 is read off as “two 1s” as the count of “1” is two in this term. 
# Hence, the next term in the sequence becomes 21.
# 21 becomes one2 and one1 (1211)

def find_next_number(number):
    result = ''
    data = (str(number))
    prev_char = temp = ''
    while len(data) > 0:
        char = data[0]
        data = data[1:]
        if prev_char == '' or char == prev_char:
            prev_char = char
            temp += char
        else:
            result += str(len(temp)) + temp[0]
            prev_char = char
            temp = char
    result += str(len(temp)) + temp[0]
    return result

def find_next_number_loops(number):
    data = str(number)
    result = []
    i = 0
    while i < len(data):
        count = 1
        while i + 1 < len(data) and data[i] == data[i+1]:
            i += 1
            count += 1
        result.append(str(count) + data[i])
        i += 1
    return ''.join(result)


A1 = 1
for i in range(10):
    temp = A1
    A1 = find_next_number_loops(A1)
    print(f"Next number for {temp} in look and say is : {A1}")

```

```python
# Spread sheet encoding

# In this lesson, we will be considering how to solve the problem of implementing a function that converts a spreadsheet column ID (i.e., “A”, “B”, “C”, …, “Z”, “AA”, etc.) to the corresponding integer. 
# For example, “A” equals 1 because it represents the first column, while “AA” equals 27 because it represents the 27th column.

# The key insight is to think of the column IDs as base 26 integers.

# In the illustration below, you can see how we represent “314” as a base 10 number.
# (3 * 10 ^2) + (1 * 10^1) + (4 * 10^0)

# let’s go ahead and represent the numbers of base 26 according to the formula above
# AA = (1*26^1) + (1*26^0) = 27

def spread_sheet_column_decode(data):
    num = 0
    count = len(data)-1
    for s in data:
        num += 26**count * (ord(s) - ord('A') + 1)
        count -= 1
    return num

print(f"{spread_sheet_column_decode('A')}")
print(f"{spread_sheet_column_decode('B')}")
print(f"{spread_sheet_column_decode('Y')}")
print(f"{spread_sheet_column_decode('Z')}")
print(f"{spread_sheet_column_decode('AA')}")
print(f"{spread_sheet_column_decode('AB')}")
print(f"{spread_sheet_column_decode('YA')}")
print(f"{spread_sheet_column_decode('ZB')}")
```

```python
# Is Palindrome

def is_palindrome_naive(data):
    data = ''.join([i for i in data if i.isalnum()]).replace(' ', '').lower()
    return data == data[::-1]

def is_palindrome_loop(data):
    i = 0
    j = len(data)-1
    while i < j:
        while not data[i].isalnum() and i < j:
            i += 1
        while not data[j].isalnum() and i < j:
            j -= 1

        if data[i].lower() != data[j].lower():
            return False
        i += 1
        j -= 1
    return True


A1 = "Was it a cat I saw?"
A2 = "radar"
A3 = "never odd or even"
A4 = "Live on time, emit no evil"
A5 = "How are you?"

print(f"{A1} is palindrmome? - {is_palindrome_naive(A1)}")
print(f"{A1} is palindrmome? - {is_palindrome_loop(A1)}")

print(f"{A2} is palindrmome? - {is_palindrome_naive(A2)}")
print(f"{A2} is palindrmome? - {is_palindrome_loop(A2)}")


print(f"{A3} is palindrmome? - {is_palindrome_naive(A3)}")
print(f"{A3} is palindrmome? - {is_palindrome_loop(A3)}")

print(f"{A4} is palindrmome? - {is_palindrome_naive(A4)}")
print(f"{A4} is palindrmome? - {is_palindrome_loop(A4)}")

print(f"{A5} is palindrmome? - {is_palindrome_naive(A5)}")
print(f"{A5} is palindrmome? - {is_palindrome_loop(A5)}")
```

```python
# Anagram

# Time complexity of O(n log(n)) and space complexity O(1)
def is_anagram_naive(input1, input2):
    input1 = input1.lower().replace(" ", "")
    input2 = input2.lower().replace(" ", "")
    return sorted(input1) == sorted(input2)

# Time complexit of O(n) but space complexity of O(n)
def is_anagram_improved(input1, input2):
    ht = dict()
    input1 = input1.lower().replace(" ", "")
    input2 = input2.lower().replace(" ", "")
    
    if (len(input1) != len(input2)):
        return False    
    for i in input1:
        if i in ht:
            ht[i] += 1
        else:
            ht[i] = 1
    for i in input2:
        if i in ht:
            ht[i] -= 1
        else:
            ht[i] = 1
    for h in ht:
        if ht[h] != 0:
            return False
    return True


A1 = "William Shakespeare"
A2 = "I am a weakish speller"
print(f"Is '{A1}' and '{A2}' an anagram? : {is_anagram_naive(A1, A2)}")
print(f"Is '{A1}' and '{A2}' an anagram? : {is_anagram_improved(A1, A2)}")


A1 = "rail safety"
A2 = "fairy tales"
print(f"Is '{A1}' and '{A2}' an anagram? : {is_anagram_naive(A1, A2)}")
print(f"Is '{A1}' and '{A2}' an anagram? : {is_anagram_improved(A1, A2)}")


A1 = "how are you"
A2 = "you are good"
print(f"Is '{A1}' and '{A2}' an anagram? : {is_anagram_naive(A1, A2)}")
print(f"Is '{A1}' and '{A2}' an anagram? : {is_anagram_improved(A1, A2)}")
```

```python
# Is Palindrom Permutation

# Given a string, write a function to check if it is a permutation of a palindrome.
# Palindrome: A word or phrase that is the same forwards and backward.
# Permutation: A rearrangement of letters.
# The palindrome does not need to be limited to just dictionary words.
# Before jumping straight to the code, let’s think about how we can approach this problem. Here is an example of a palindrome:

# T A C O C A T
# Every charater has a duplicate in this odd-length string except the one chracter in the middle

# the idea we get is that a string that has an even length must have all even counts of characters, 
# while strings that have an odd length must have exactly one character with an odd count. 
# An even-length-ed string can’t have an odd number of exactly one character; otherwise, it wouldn’t be even. 
# This is true since an odd number plus any set of even numbers will yield an odd number.

def is_palindram_permutation(input1):
    input1 = input1.replace(" ", "").lower()

    ht = dict()
    for i in input1:
        if i in ht:
            ht[i] += 1
        else:
            ht[i] = 1

    odd_count = 0
    for k, v in ht.items():
        if v%2 != 0 and odd_count == 0:
            odd_count += 1
        elif v%2 != 0 and odd_count !=0:
            return False
    return True

str1 = "Tact Coa"
str2 = "This is not a palindrome permutation"

print(f"{str1} is Palindrome Permutation: {is_palindram_permutation(str1)}")
print(f"{str2} is Palindrome Permutation: {is_palindram_permutation(str2)}")
```

```python
# Check one string could be a permutation of another string
# Given two strings, write a function to determine if one is a permutation of the other.

# For example
# str1 = "google" & str2 = "oogleg" is permutation of one another
# str2 = "not" and str2 = "top" not a permutation of one another

def is_permutation_string(str1, str2):
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()
    if len(str1) != len(str2):
        return False
    return sorted(str1) == sorted(str2)

def is_permutation_string_logic(str1, str2):
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()
    if len(str1) != len(str2):
        return False
    
    ht = dict()
    for i in str1:
        if i in ht:
            ht[i] += 1
        else:
            ht[i] = 1
    for i in str2:
        if i in ht:
            ht[i] -= 1
        else:
            return False
    return all(value == 0 for value in ht.values())




str1 = 'google'
str2 = 'gleoog'
print(f"'{str1}' and '{str2}' - are they permutation of each other - {is_permutation_string(str1, str2)}")
print(f"'{str1}' and '{str2}' - are they permutation of each other - {is_permutation_string_logic(str1, str2)}")

str1 = 'top'
str2 = 'not'
print(f"'{str1}' and '{str2}' - are they permutation of each other - {is_permutation_string(str1, str2)}")
print(f"'{str1}' and '{str2}' - are they permutation of each other - {is_permutation_string_logic(str1, str2)}")
```

```python
# Is Unique?

# Your task is to implement an algorithm to determine if a string has all unique characters.
# Assume that the input string will only contain alphabets or spaces.

def is_unique_string(str1):
    str1 = str1.replace(" ", "").lower()
    unique_lst = []
    for i in str1:
        if i not in unique_lst:
            unique_lst.append(i)
        else:
            return False
    return True

def is_unique_string_naive(str1):
    return len(set(str1)) == len(str1)

str1 = "abcdefghijklm"
print(f"'{str1}' has all unique chars : {is_unique_string(str1)}")
print(f"'{str1}' has all unique chars : {is_unique_string_naive(str1)}")


str1 = "not unique"
print(f"'{str1}' has all unique chars : {is_unique_string(str1)}")
print(f"'{str1}' has all unique chars : {is_unique_string_naive(str1)}")
```

```python
# Integer to string
# You are given some integer as input, (i.e. … -3, -2, -1, 0, 1, 2, 3 …) and you have to convert the integer you are given to a string. 
# Examples:
# 123 = '123'
# -123 = '-123'

def int_2_str(number):
    is_negative = True if number < 0 else False

    number = abs(number)
    result = ''
    if number == 0:
        result = "0"
    else:
        while number > 0:
            result += chr(ord('0') + number %10)
            number = number //10
        result = result[::-1]

    result = ("-" if is_negative == True else "") + result
    return result

n = 123
print(f"{n}'s string value is : {int_2_str(n)}")

n = -123
print(f"{n}'s string value is : {int_2_str(n)}")
```

```python
# String to integer

# you are given some numeric string as input. You have to convert the string you are given to an integer.
# Do not make use of the built-in int function.

# Example:
# "123" -> 123
# "-12332" -> -12332

def string_2_int(str1):
    is_negative = False if str1.find('-') == -1 else True
    str1 = str1.replace('-', '')
    pos = 0
    result = 0
    for i in str1[::-1]:
        result += 10**pos * (ord(i)- ord('0'))
        pos += 1
    result *= -1 if is_negative else 1
    return result

str1 = '123'
print(f"'{str1}' integer value is {string_2_int(str1)}")

str1 = '-12332'
print(f"'{str1}' integer value is {string_2_int(str1)}")
```
